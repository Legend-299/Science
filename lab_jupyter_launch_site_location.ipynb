```python
# Import distance calculation library
from math import sin, cos, sqrt, atan2, radians

# Function to calculate distance between two points on Earth
def calculate_distance(lat1, lon1, lat2, lon2):
    # Approximate radius of Earth in km
    R = 6373.0
    
    # Convert coordinates to radians
    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)
    
    # Calculate differences
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    
    # Haversine formula
    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    
    distance = R * c
    return distance

# Get unique launch sites
launch_sites = spacex_df[['Launch Site', 'Lat', 'Long']].drop_duplicates()

# Calculate distances between all launch sites
print("Distances between launch sites (in km):")
print("-" * 50)

for idx1, site1 in launch_sites.iterrows():
    for idx2, site2 in launch_sites.iterrows():
        if site1['Launch Site'] != site2['Launch Site']:
            distance = calculate_distance(site1['Lat'], site1['Long'], 
                                        site2['Lat'], site2['Long'])
            print(f"{site1['Launch Site']} to {site2['Launch Site']}: {distance:.2f} km")
    print()

# Define some proximity coordinates (example: coastline, cities)
proximities = {
    'Coastline': (28.56, -80.57),  # Example coordinate near coast
    'Cape Canaveral city': (28.3922, -80.6077),
    'Port Canaveral': (28.4158, -80.6203)
}

# Calculate distance from each launch site to proximities
print("\nDistances to nearby features:")
print("-" * 50)

for idx, site in launch_sites.iterrows():
    print(f"\n{site['Launch Site']}:")
    for feature_name, (lat, lon) in proximities.items():
        distance = calculate_distance(site['Lat'], site['Long'], lat, lon)
        print(f"  Distance to {feature_name}: {distance:.2f} km")
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
# Define the coastline coordinate (example provided)
coastline_lat = 28.56367
coastline_lon = -80.57163

# Calculate distance from each launch site to the coastline
print("Distance from each launch site to the coastline:")
print("-" * 60)

for idx, site in launch_sites.iterrows():
    distance_coastline = calculate_distance(site['Lat'], site['Long'], 
                                          coastline_lat, coastline_lon)
    print(f"{site['Launch Site']}: {distance_coastline:.2f} km to coastline")

# If you want to find multiple coastline points and find the closest one:
# Define multiple coastline points (these are approximate coordinates)
coastline_points = [
    {'name': 'East Coast Point 1', 'lat': 28.56367, 'lon': -80.57163},
    {'name': 'East Coast Point 2', 'lat': 28.45, 'lon': -80.52},
    {'name': 'East Coast Point 3', 'lat': 28.65, 'lon': -80.60},
    {'name': 'Port Canaveral', 'lat': 28.4158, 'lon': -80.6103},
    {'name': 'Cocoa Beach', 'lat': 28.3200, 'lon': -80.6076}
]

# Find the closest coastline point for each launch site
print("\n\nClosest coastline point for each launch site:")
print("-" * 60)

for idx, site in launch_sites.iterrows():
    min_distance = float('inf')
    closest_point = None
    
    for point in coastline_points:
        distance = calculate_distance(site['Lat'], site['Long'], 
                                    point['lat'], point['lon'])
        if distance < min_distance:
            min_distance = distance
            closest_point = point
    
    print(f"{site['Launch Site']}:")
    print(f"  Closest point: {closest_point['name']}")
    print(f"  Coordinates: Lat: {closest_point['lat']}, Lon: {closest_point['lon']}")
    print(f"  Distance: {min_distance:.2f} km\n")

# Visualize on map
coastline_map = folium.Map(location=[28.5, -80.6], zoom_start=10)

# Add launch site markers
for idx, site in launch_sites.iterrows():
    folium.Marker(
        location=[site['Lat'], site['Long']],
        popup=site['Launch Site'],
        icon=folium.Icon(color='red', icon='star')
    ).add_to(coastline_map)

# Add coastline markers
for point in coastline_points:
    folium.Marker(
        location=[point['lat'], point['lon']],
        popup=point['name'],
        icon=folium.Icon(color='blue', icon='info-sign')
    ).add_to(coastline_map)

# Add lines from launch sites to nearest coastline
for idx, site in launch_sites.iterrows():
    # Find closest coastline point again
    min_distance = float('inf')
    closest_point = None
    
    for point in coastline_points:
        distance = calculate_distance(site['Lat'], site['Long'], 
                                    point['lat'], point['lon'])
        if distance < min_distance:
            min_distance = distance
            closest_point = point
    
    # Draw line from launch site to closest coastline
    folium.PolyLine(
        locations=[[site['Lat'], site['Long']], 
                   [closest_point['lat'], closest_point['lon']]],
        color='green',
        weight=2,
        opacity=0.6,
        popup=f"Distance: {min_distance:.2f} km"
    ).add_to(coastline_map)

coastline_map
```

```python
# Define nearby points of interest (cities, railways, highways)
points_of_interest = [
    {'name': 'Orlando', 'type': 'city', 'lat': 28.5383, 'lon': -81.3792},
    {'name': 'Miami', 'type': 'city', 'lat': 25.7617, 'lon': -80.1918},
    {'name': 'Tampa', 'type': 'city', 'lat': 27.9506, 'lon': -82.4572},
    {'name': 'Jacksonville', 'type': 'city', 'lat': 30.3322, 'lon': -81.6557},
    {'name': 'Cocoa Beach', 'type': 'city', 'lat': 28.3200, 'lon': -80.6076},
    {'name': 'Port Canaveral', 'type': 'port', 'lat': 28.4104, 'lon': -80.6328},
    {'name': 'Highway I-95', 'type': 'highway', 'lat': 28.5618, 'lon': -80.8003},
    {'name': 'Highway 528', 'type': 'highway', 'lat': 28.4213, 'lon': -80.7526},
    {'name': 'Florida East Coast Railway', 'type': 'railway', 'lat': 28.5221, 'lon': -80.7890}
]

# Create a new map
poi_map = folium.Map(location=[28.5, -80.6], zoom_start=8)

# Add launch site markers
for idx, site in launch_sites.iterrows():
    folium.Marker(
        location=[site['Lat'], site['Long']],
        popup=site['Launch Site'],
        icon=folium.Icon(color='red', icon='rocket')
    ).add_to(poi_map)

# Add points of interest markers
for poi in points_of_interest:
    # Choose icon based on type
    if poi['type'] == 'city':
        icon_type = 'home'
        color = 'blue'
    elif poi['type'] == 'port':
        icon_type = 'ship'
        color = 'purple'
    elif poi['type'] == 'highway':
        icon_type = 'road'
        color = 'green'
    elif poi['type'] == 'railway':
        icon_type = 'train'
        color = 'orange'
    else:
        icon_type = 'info-sign'
        color = 'gray'
    
    folium.Marker(
        location=[poi['lat'], poi['lon']],
        popup=poi['name'],
        icon=folium.Icon(color=color, icon=icon_type)
    ).add_to(poi_map)

# For each launch site, find closest point of each type and draw lines
for idx, site in launch_sites.iterrows():
    # Group POIs by type
    poi_by_type = {}
    for poi in points_of_interest:
        if poi['type'] not in poi_by_type:
            poi_by_type[poi['type']] = []
        poi_by_type[poi['type']].append(poi)
    
    # Find closest of each type
    for poi_type, poi_list in poi_by_type.items():
        closest_poi = None
        min_distance = float('inf')
        
        for poi in poi_list:
            distance = calculate_distance(site['Lat'], site['Long'], 
                                        poi['lat'], poi['lon'])
            if distance < min_distance:
                min_distance = distance
                closest_poi = poi
        
        # Create line to closest POI of this type
        if closest_poi:
            # Choose line color based on type
            if poi_type == 'city':
                line_color = 'blue'
            elif poi_type == 'port':
                line_color = 'purple'
            elif poi_type == 'highway':
                line_color = 'green'
            elif poi_type == 'railway':
                line_color = 'orange'
            else:
                line_color = 'gray'
            
            # Create line
            line = folium.PolyLine(
                locations=[
                    [site['Lat'], site['Long']],
                    [closest_poi['lat'], closest_poi['lon']]
                ],
                color=line_color,
                weight=2,
                opacity=0.7,
                popup=f"{site['Launch Site']} to {closest_poi['name']}: {min_distance:.2f} km"
            )
            poi_map.add_child(line)
            
            # Add distance marker
            distance_marker = folium.Marker(
                location=[(site['Lat'] + closest_poi['lat'])/2, 
                          (site['Long'] + closest_poi['lon'])/2],
                icon=DivIcon(
                    icon_size=(60, 20),
                    icon_anchor=(30, 10),
                    html=f'<div style="font-size: 10px; background-color: white; border: 1px solid {line_color}; padding: 2px; border-radius: 3px;">{min_distance:.1f} km</div>'
                )
            )
            poi_map.add_child(distance_marker)

# Display the map
poi_map
```
